```python
%sql
CREATE SCHEMA IF NOT EXISTS healthcare.governance 
```

```python
%sql
CREATE OR REPLACE FUNCTION healthcare.governance.mask_ssn(v STRING)
RETURNS STRING
RETURN CASE
    WHEN current_user() = 'doctor@company.com' THEN v
    WHEN current_user() = 'billing@company.com' THEN concat('XXX-XX-', right(v, 4))
    ELSE concat('XXX-XX-', right(v, 4))
END;
```

```python
%sql
ALTER TABLE healthcare.gold.dim_patient
ALTER COLUMN ssn
SET MASK healthcare.governance.mask_ssn;
```

```python
%sql
CREATE OR REPLACE FUNCTION healthcare.governance.mask_mrn(v STRING)
RETURNS STRING
RETURN CASE
    WHEN current_user() = 'doctor@company.com' THEN v
    ELSE '***REDACTED***'
END;
```

```python
%sql
ALTER TABLE healthcare.gold.dim_patient
ALTER COLUMN mrn
SET MASK healthcare.governance.mask_mrn;
```
